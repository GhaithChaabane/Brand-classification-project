```python
def replace_non_existent_image_paths(df, column_name):
  """Replaces non-existing image paths in a pandas DataFrame column with an empty string."""
  for i in range(len(df)):
    path = df.loc[i, column_name]

    if not os.path.isfile(path):
        df.loc[i, column_name] = ""

  return df
```

# Create Dataloaders

```python

# Define constants
BATCH_SIZE = 1792
NUM_CLASSES = 6077 
MAX_TEXT_LENGTH = 50









def custom_data_generator(image_data,text_data, labels, batch_size):
    num_samples = len(labels)
    while True:
        indices = np.arange(num_samples)
        np.random.shuffle(indices)  # Shuffle the indices for each epoch
        for start_idx in range(0, num_samples, batch_size):
            end_idx = min(start_idx + batch_size, num_samples)
            batch_indices = indices[start_idx:end_idx]
            batch_images=[]
            for path in np.array(image_data)[batch_indices]:
                if path == '':
                    image_array = np.zeros((desired_height, desired_width, desired_channels), dtype=np.uint8)  # Replace dimensions as needed
                else:
                    image = cv2.imdecode(np.fromfile(path, dtype=np.uint8),cv2.IMREAD_UNCHANGED)  # Read image using OpenCV
                    image_array = cv2.resize(image, (desired_width, desired_height))  # Resize if needed
                    image_array=image_array / 60.0  # Apply MobileNetV2 preprocessing
                batch_images.append(image_array)            
            batch_texts = text_data[batch_indices]
            batch_labels=enc.transform(labels.values.reshape(-1,1)[batch_indices]).toarray()

            yield [np.array(batch_images),batch_texts], batch_labels



steps_per_epoch = X_train.shape[0] // BATCH_SIZE
if X_train.shape[0] % BATCH_SIZE:
    steps_per_epoch += 1

# Get number of validation steps.
validation_steps = X_test.shape[0] // BATCH_SIZE
if X_test.shape[0] % BATCH_SIZE:
    validation_steps += 1
# Create the data generator
```

```python
data_generator = custom_data_generator(image_url, X_train_, y_train, BATCH_SIZE)
test_generator = custom_data_generator(image_test, X_test_, y_test, BATCH_SIZE)
```

```python
my_callbacks = [
    tf.keras.callbacks.EarlyStopping(patience=2),
    tf.keras.callbacks.ModelCheckpoint(filepath='model.{epoch:02d}-{val_loss:.2f}.h5'),
    tf.keras.callbacks.TensorBoard(log_dir='New folder'),
]
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit_generator(data_generator,steps_per_epoch,validation_data=test_generator,validation_steps=validation_steps, epochs=1, callbacks=my_callbacks,shuffle=True)
```
